```python
import numpy as np
from PIL import Image
from os import path
import matplotlib.pyplot as plt
import random
import matplotlib.pylab as pylab
```

```python
from wordcloud import WordCloud, STOPWORDS
def grey_color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)

d = path.dirname('Wong_et_al-2017-Groundwater.txt')
d

# Read the whole text.
text = open(path.join(d, 'Wong_et_al-2017-Groundwater.txt')).read()
```

```python
def replaceText(text):
    # preprocessing the text a little bit
    text = text.replace("viruses", "virus")
    text = text.replace("HAdV", "adenovirus")
    text = text.replace("samples", "sample")
    text = text.replace("cowpat", "fecal")
    text = text.replace("feces", "fecal")
    text = text.replace("fecals", "fecal")
    text = text.replace("Wong", "wong")
    text = text.replace("difference", "different")
    text = text.replace("increased", "increase")
    text = text.replace("community structure", "microbiome")
    text = text.replace("bacteria", "microbiome")
    text = text.replace("studies", "study")
    text = text.replace("levels", "level")
    text = text.replace("results", "result")
    text = text.replace("NV", "norovirus")
    text = text.replace("EV", "enterovirus")
    text = text.replace("HPyV", "polyomavirus")
    text = text.replace("used", "use")
    text = text.replace("Environ", "environment")
    text = text.replace("Figure", "figure")
    text = text.replace("Table", "table")
    text = text.replace("ZP", "Zeta potential")
    text = text.replace("pHiep", "isoelectric point")
    text = text.replace("DH", "hydrodynameter")
    text = text.replace("BTC", "breakthrough curve")
    return text


# adding movie script specific stopwords
stopwords = set(STOPWORDS)
stopwords.add("int")
stopwords.add("ext")
stopwords.add("et")
stopwords.add("al")
stopwords.add("virus")
stopwords.add("NA")
stopwords.add("study")
stopwords.add("since")
stopwords.add("using")
stopwords.add("different")
stopwords.add("one")
stopwords.add("RA")
stopwords.add("S2")
stopwords.add("S1")
stopwords.add("U1")
stopwords.add("U2")
stopwords.add("et")
stopwords.add("al")
stopwords.add("wong")
stopwords.add("day")
stopwords.add("sample")
stopwords.add("fecal")
stopwords.add("Figure")
stopwords.add("increase")
stopwords.add("level")
stopwords.add("also")
stopwords.add("Supplementary")
stopwords.add("due")
stopwords.add('sample')
stopwords.add("Appl")
stopwords.add("result")
stopwords.add("MAD")
stopwords.add("previous")
stopwords.add("table")
stopwords.add("class")
stopwords.add("time")
stopwords.add("Microbiol")
stopwords.add("use")
stopwords.add("However")
stopwords.add("figure")
stopwords.add("concentration")
stopwords.add("standard")
stopwords.add("mM")
stopwords.add("1mM")
stopwords.add("point")
stopwords.add("condition")
stopwords.add("ml")
stopwords.add("e.ect")
stopwords.add("XXX")
```

```python
# Read the whole text.
text1 = open(path.join(d, 'Wong_et_al-2017-Groundwater.txt')).read()
text1 = replaceText(text1)

wc = WordCloud(max_words=1000,  stopwords=stopwords, margin=10,
               random_state=1).generate(text1)
# store default colored image
default_colors = wc.to_array()
plt.figure(figsize=(15, 15))
plt.imshow(default_colors, interpolation="bilinear")
plt.title("Applying Quantitative Molecular Tools for Virus Transport Studies: Opportunities and Challenges (2017)", 
          fontsize=15)
plt.axis("off")
plt.savefig('2017.png')
```
