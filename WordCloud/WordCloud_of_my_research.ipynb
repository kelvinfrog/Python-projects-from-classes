```python
# Read the whole text.
text1 = open(path.join(d, 'Wong_et_al-2017-Groundwater.txt')).read()
text1 = replaceText(text1)

wc = WordCloud(max_words=1000,  stopwords=stopwords, margin=10,
               random_state=1).generate(text1)
# store default colored image
default_colors = wc.to_array()
plt.figure(figsize=(15, 15))
plt.imshow(default_colors, interpolation="bilinear")
plt.title("Applying Quantitative Molecular Tools for Virus Transport Studies: Opportunities and Challenges (2017)", 
          fontsize=15)
plt.axis("off")
plt.savefig('2017.png')
```
